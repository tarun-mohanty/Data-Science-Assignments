Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a type of plot used to display the relationship or correlation between two numerical variables. In a scatter plot, each data point is represented by a dot on a two-dimensional coordinate system.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function is used to create a grid of subplots within a single figure. 

```python
plt.subplot(2,2,1)
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x1,y1)

plt.subplot(2,2,2)
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x2,y2)

plt.subplot(2,2,3)
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x3,y3)

plt.subplot(2,2,4)
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x4,y4)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart, is a type of plot that uses rectangular bars to represent the values of categorical or discrete variables. It is used because they provide a clear and intuitive way to represent and interpret categorical or discrete data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a continuous variable. It displays the summary statistics of the data, such as the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values, along with any outliers.

It is used because it provide a compact and informative way to represent the distribution of continuous variables, making it easier to compare groups, identify outliers, and understand the overall characteristics of the data.

```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
plt.show()
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.show()
```
